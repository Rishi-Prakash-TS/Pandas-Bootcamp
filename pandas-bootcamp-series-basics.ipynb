```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
#We start with series and then escalate to dataframe in the future
from string import ascii_lowercase
from string import ascii_uppercase
#Creating series with aphabets
alphabet= pd.Series(data=list(ascii_lowercase), index=list(ascii_uppercase))
alphabet.head()
```

```python
#We can add prefix and suffix to series using the following funtion
alphabet= alphabet.add_prefix('Label_')
```

```python
alphabet.head()
```

```python
#Instead of using [] for indexing we can also call values using dot notations in series
alphabet.Label_V
```

# Boolean Masks and .loc, .iloc indexer

```python
#For indexing by labels we can also use .loc as it is faster and product specific
alphabet.loc['Label_B':'Label_F']
```

```python
#We can also specify boolean values in index. Note: The values must match the count of total items
alphabet.loc[[True for i in range(26) ]]
```

```python
#Setting alternate values as True
alphabet.loc[[True if i%2==0 else False for i in range(26) ]]
```

```python
#.iloc is used to index based on position
alphabet.iloc[[1,2,6,8]]
```

```python
alphabet.iloc[3:7]
```

# Collables

```python
#It is the same as using [], .loc or .iloc with the addition of custamizable index
#Eg Every fifth item

alphabet.iloc[lambda x: [True if i%5==0 else False for i in range(x.size)]]
```

```python
#Get to index. Get has the option of specifying not available values.
#Get can take in both positions and label
alphabet.get("Label_V_Sike") == None
```

```python
alphabet.get("Sike",default="Got you")
```
